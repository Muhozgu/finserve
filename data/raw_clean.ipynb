```python
def clean_phone(value):
    s = str(value).strip().lower()

    # Handle missing/junk values
    if s in ['nan', 'none', '', 'null', '-']:
        return None

    # Strip extension (x1234)
    s = re.split(r'x\d+', s)[0]

    # Keep only digits
    digits = re.sub(r'\D', '', s)

    # Normalize to +1XXXXXXXXXX
    if len(digits) == 11 and digits.startswith('1'):
        digits = digits[1:]
    elif len(digits) == 13 and digits.startswith('001'):
        digits = digits[3:]

    if len(digits) == 10:
        return '+1' + digits

    return None  # couldn't normalize -> leave as missing for now
```

```python
# Apply and overwrite the column directly
df['phone'] = df['phone'].apply(clean_phone)
```

```python
def clean_person_home_ownership(value):
  s = str(value).upper()

  return s

df['person_home_ownership'] = df['person_home_ownership'].apply(clean_person_home_ownership)
```
